```python
#List Creation
##a.Create a list named age_list with five integer elements.

age_list = [24,25,27,28,29]
print("age_list:", age_list)

##b.Create a list named name_list with five string elements.
name_list = ["Madhu","Magues","Mekha","Deiva","Subash"]
print("name_list:",name_list)

#List Operations / Modifications:
##a. Append the string "Yazhini" to name_list.

name_list.append("Yazhini")
print(name_list)

##b. Insert the element 26 at index 2 in age_list.
age_list.insert(2,26)
print(age_list)

##c.Remove the string "Yazhini" from name_list.
name_list.remove("Yazhini")
print(name_list)

##d.Pop the last element from age_list
age_list.pop(5)
print(age_list)

##e.Extend the age_list with additional ages [31,30,32].
age_list.extend([31,30,32])
print(age_list)

##f.Sort age_list in descending order.
age_list.sort(reverse=True)
print(age_list)

##g. Find the max age, min age, and sum of all ages from age_list.
print("Max age:",max(age_list))
print("Min age:",min(age_list))
print("Sum of all ages:",sum(age_list))

#Accessing List Elements
##a.Print the first element of name_list.
print(name_list[0])

##b.Print the last element of name_list.
print(name_list[-1])

##c.Print the elements from index 2 to index 4 in name_list.
print(name_list[2:5])

##d.Print the elements of name_list in reverse order.
print(name_list[::-1])
```

```python
#Dictionary (Creation, Modification and Access):
##a. Create a dictionary named student_marks that maps the names of five students to their marks (use a scale of 0 to 100).
student_marks = {
    "Madhu":85,
    "Magues":90,
    "Mekha":91,
    "Deiva":89,
    "Subash":97
}
print("Students marks:",student_marks)

##b. Access and print the mark of a specific student of your choice.
print("Mekha Marks:",student_marks['Mekha'])

##c.Add a new student, "Janani," with a mark of 80 to the student_marks dictionary
student_marks["Janani"]=80
print(student_marks)

##d. Update the mark of any one older student to 82.
student_marks["Madhu"]= 82
print(student_marks)

##c.Use the keys(), values(), and items() methods to print all keys, values, and key-value pairs in the student_marks dictionary.

print("Keys:",student_marks.keys())
print("Values:",student_marks.values())
print("Items:",student_marks.items())
```
