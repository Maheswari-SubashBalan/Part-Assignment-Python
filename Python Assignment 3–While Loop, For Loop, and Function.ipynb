```python
#Task 1: While Loop & Control Statements -Number Guessing Game
##1. Set Up the Game:
import random
# Generate random number between 1 and 10
secret_number = random.randint(1,10)

##2. Prompt the User
attempts = 3

##3.Implement the Guessing Logic:
while attempts > 0:
  guess=int(input("Guess the number(between 1 and 10):"))

  #If the guess is out of the valid range
  if guess <1 or guess>10:
    print("Your guess is out of range. Please guess a number between 1 and 10.")
    continue #skip the rest of the loop after informing the user if the guess is out of range
  attempts-=1
  if guess < secret_number:
    print("Too low. Try again")
  elif guess > secret_number:
    print("Too low. Try again")
  else:
    print("Congratulations! You guessed the correct number")
    break #exit the loop when the guess is correct
else:
  print("Better luck next time!")


```

```python
#Task 2:- For Loop - Multiplication Table Generator
##Prompt user for input
number = int(input("enter a number for which they want to generate a multiplication table :"))

#Generate the Multiplication Table:
for i in range(1, 11):
    result = number * i
    print(f"{number} x {i} = {result}")
```

```python
#Task 3: Function- BMI Calculator
##Define the function to calculate BMI

def calculate_bmi(weight,height):
  bmi = weight / (height ** 2)
  return bmi

weight = float(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))

bmi = calculate_bmi(weight,height)
print(f"Your BMI is: {bmi:.2f}")
```
